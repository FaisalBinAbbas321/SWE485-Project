# Phase 1 – Data Exploration (Healthcare – Stroke Risk)


## Goal & Source
- Goal and purpose: predict or give advice about stroke risk and The purpose of this dataset is to help predict the **risk of stroke** based on patient health information such as age, glucose level, BMI, smoking status, and medical history.  


- Dataset: Stroke Prediction Dataset.
- Source: https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset
- Why: small, clear dataset with health features and target column and its relates to healthcare, which is important and interesting.


```python
#The Dataset
import pandas

dataset_link = "https://raw.githubusercontent.com/FaisalBinAbbas321/SWE485-Project/main/Dataset/Stroke-data.csv"

stroke_data = pandas.read_csv(dataset_link)

stroke_data.head()
```

```python

#General Information
print("Number of rows and columns:")
print(stroke_data.shape)

print("\nColumn names:")
print(list(stroke_data.columns))

print("\nColumn information:")
print(stroke_data.info())


```

 Description of the target variable:

Target variable: stroke      
0 = No stroke      
1 = Stroke occurred

```python
#Summary and visualization:

print("Statistical Summary:")
print(stroke_data.describe())

print("Missing values per column:")
print(stroke_data.isnull().sum())

print("Count of each class in target column:")
print(stroke_data['stroke'].value_counts())
# 0 means no Stroke and 1 means Stroke occured

import matplotlib.pyplot as charts

# Age
stroke_data['age'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black', title='Age Distribution')
charts.xlabel("Age")
charts.ylabel("Count")
charts.show()

# Glucose Level
stroke_data['avg_glucose_level'].plot(kind='hist', bins=30, color='orange', edgecolor='black', title='Average Glucose Level Distribution')
charts.xlabel("Average Glucose Level")
charts.ylabel("Count")
charts.show()


```

```python
#Preprocessing Techniques:

# Check before cleaning
print("Missing values before cleaning:")
print(stroke_data.isnull().sum())

# Fill missing  with average
stroke_data['bmi'] = stroke_data['bmi'].fillna(stroke_data['bmi'].mean())

# Convert text columns into numbers
stroke_data = pandas.get_dummies(stroke_data, drop_first=True)

# Check after cleaning
print("Missing values after cleaning:")
print(stroke_data.isnull().sum())
```
